```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException

driver1 = webdriver.Chrome()
driver1.get("https://www.amazon.in/")

# 2. Enter the product name in the search field
product_name = str(input("Please enter the product's name: "))

search1 = driver1.find_element(By.XPATH, "/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[2]/div[1]/input")
search1.send_keys(product_name)

search_button = driver1.find_element(By.XPATH, "/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[3]/div/span/input")
search_button.click()

# Initialize empty lists
submit = []
product_url = []

# Iterate over pages
for page in range(0, 2):
    products = driver1.find_elements(By.XPATH, '//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for product in products:
        product_url.append(product.get_attribute('href'))

    try:
        submit_page = driver1.find_element(By.XPATH, '/html/body/div[1]/div[1]/div[1]/div[1]/div/span[1]/div[1]/div[63]/div/div/span/a[3]')
        submit.append(submit_page.get_attribute('href'))
    except NoSuchElementException:
        submit = f"https://www.amazon.in/s?k=guitar&page={page+2}&crid=34NTE1DJMA24H&qid=1706983933&sprefix=%2Caps%2C107&ref=sr_pg_{page+1}"
    driver1.get(submit)
    








```

```python
brand = []
name = []
price = []
return_exchange = []
delivery_date = []
availability = []

for a in product_url:  
    driver1.get(a)
    time.sleep(5)
    
    # Initialize placeholder values
    brand_value = '-'
    name_value = '-'
    price_value = '-'
    return_exchange_value = '-'
    delivery_date_value = '-'
    availability_value = '-'

    try:
        brand_driver = driver1.find_element(By.XPATH, '/html/body/div[2]/div/div[5]/div[3]/div[4]/div[48]/div/table/tbody/tr[1]/td[2]/span')
        brand_value = brand_driver.text

        name_driver = driver1.find_element(By.XPATH, '/html/body/div[2]/div/div[5]/div[3]/div[4]/div[1]/div/h1/span')
        name_value = name_driver.text

        price_driver = driver1.find_element(By.XPATH, '//span[@class="a-price aok-align-center reinventPricePriceToPayMargin priceToPay"]/span[2]/span[2]')
        price_value = price_driver.text

        return_exchange_driver = driver1.find_element(By.XPATH, '/html/body/div[2]/div/div[5]/div[3]/div[4]/div[24]/div[2]/div/div/div/div[2]/div/ol/li[3]/div/span/div[2]/span')
        return_exchange_value = return_exchange_driver.text

        delivery_date_driver = driver1.find_element(By.XPATH, '/html/body/div[2]/div/div[5]/div[3]/div[1]/div[3]/div/div[1]/div/div/div/form/div/div/div/div/div[4]/div/div[3]/div[10]/div[1]/div/div/div/span/span[1]')
        delivery_date_value = delivery_date_driver.text

        availability_driver = driver1.find_element(By.XPATH, '/html/body/div[2]/div/div[5]/div[3]/div[1]/div[3]/div/div[1]/div/div/div/form/div/div/div/div/div[4]/div/div[5]/div/div[1]/span')
        availability_value = availability_driver.text

    except NoSuchElementException:
        pass  # No need to append placeholder values here

    # Append values to lists outside of the try-except block
    brand.append(brand_value)
    name.append(name_value)
    price.append(price_value)
    return_exchange.append(return_exchange_value)
    delivery_date.append(delivery_date_value)
    availability.append(availability_value)

df = pd.DataFrame({'Product title': name, 'Product brand': brand, 'Price': price, 'Return': return_exchange, 'Delivery': delivery_date, 'Availability': availability})
df.to_csv('output.csv', index=False)
```
