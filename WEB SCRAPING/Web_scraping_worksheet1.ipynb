Answer 2.
The url provided by the assignment worksheet redirects to a NOT FOUND page, therefore I have used the home page of the website to complete the assignment.

```python
import re
page2 = requests.get('https://presidentofindia.nic.in/')
soup2 = BeautifulSoup(page2.content)
Names2 = []
Terms2 = []
for i in soup2.find_all('h3', class_ = "wgl-services_title"):
    Names2.append(i.text)


for a in soup2.find_all('div', class_ = "wgl-services_text"):
    Terms2.append(a.text)
      
date_pattern = re.compile(r'\b(\d{1,2} \w+ \d{4})\b')

dates = []
for b in Terms2:
    dates.append(date_pattern.findall(b))
    

result2 = pd.DataFrame({'Names': Names2, 'Term Dates': dates})


result2
```

```python
page3a = requests.get('https://www.icc-cricket.com/rankings/team-rankings/mens/odi')
soup3a = BeautifulSoup(page3a.content)

page3b = requests.get('https://www.icc-cricket.com/rankings/batting/mens/odi')
soup3b = BeautifulSoup(page3b.content)

page3c = requests.get('https://www.icc-cricket.com/rankings/bowling/mens/odi')
soup3c = BeautifulSoup(page3c.content)

Matches3a = []
Points3a = []
Rating3a = []
Teams3b = []
Rating3b = []
Teams3c = []
Rating3c = []

for a in soup3a.find_all('div', class_ = "si-table-data si-matches"):
    Matches3a.append(a.text)


for b in soup3a.find_all('div', class_ = "si-table-data si-pts"):
    Points3a.append(b.text)


for c in soup3a.find_all('div', class_ = "si-table-data si-rating"):
    Rating3a.append(c.text)


for d in soup3b.find_all('div', class_ = "si-table-data si-team"):
    Teams3b.append(d.text)

for c in soup3b.find_all('div', class_ = "si-table-data si-rating"):
    Rating3b.append(c.text)


for d in soup3c.find_all('div', class_ = "si-table-data si-team"):
    Teams3c.append(d.text)

for e in soup3c.find_all('div', class_ = "si-table-data si-rating"):
    Rating3c.append(e.text)




result3a = pd.DataFrame({'Matches': Matches3a, 'Points': Points3a, 'Ratings': Rating3a})

result3b = pd.DataFrame({'Teams': Teams3b, 'Ratings': Rating3b})

result3c = pd.DataFrame({'Teams': Teams3c, 'Ratings': Rating3c})


result3a.head(10)
result3b.head(10)
result3c.head(10)
```

```python
page5 = requests.get('https://www.cnbc.com/world/?region=world')
soup5 = BeautifulSoup(page5.content)

Headline5 = []
Time5 = []
Link5 = []

for i in soup5.find_all('li', class_ = "LatestNews-item"):
    Headline5.append(i.text)

for a in soup5.find_all('span', class_ = "LatestNews-wrapper"):
    Time5.append(a.text)
    
for i in soup5.find_all('a', class_ = "LatestNews-headline"):
    Link5.append(i.text)



result5 = pd.DataFrame({'Headlines': Headline5, 'Time': Time5, 'News Link': Link5})


result5

```

```python
page7 = requests.get('https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles')
soup7 = BeautifulSoup(page7.content)


Restaurantname7 = []
Cuisine7 = []
Location7 = []
Ratings7 = []
ImageURL7 = []

for i in soup7.find_all('h2', class_ = "sc-1qrq3sd-1 gRGSUS sc-1nmom32-0 sc-1nmom32-1 btcbYu goSKRg"):
    Restaurantname7.append(i.text)

for a in soup7.find_all('span', class_ = "sc-1w3fpd7-0 dnCnAO"):
    Cuisine7.append(a.text)
    
for b in soup7.find_all('span', class_ = "sc-1thf9ly-2 dvggWt"):
    Location7.append(b.text)

for c in soup7.find_all('a', class_ = "sc-5smygv-0 fIXTHm"):
    Ratings7.append(c.text)

for d in soup7.find_all('a', class_ = "sc-5smygv-0 fIXTHm"):
    ImageURL7.append(d['data-src'])


result7 = pd.DataFrame({'Restaurant name': Restaurantname7, 'Cuisine': Cuisine7, 'Location': Location7 , 'Ratings': Ratings7, 'Image URL': ImageURL7})


result7
```
