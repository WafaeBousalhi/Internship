```python
# 1.Fetch the webpage 

driver = webdriver.Chrome()
driver.get("https://www.shine.com/")

# 2.Enter the data analyst in the job title field, and Bangalore in the location field

search = driver.find_element(By.XPATH, "/html/body/div/header/div[3]/div/div/div[1]/div/input")
search.click()

time.sleep(30)

close = driver.find_element(By.CLASS_NAME, "btn-close ")
close.click()

designation = driver.find_element(By.XPATH, "/html/body/div/div[4]/div/div[2]/div[2]/div/form/div/div[1]/ul/li[1]/div/input")
designation.send_keys("Data Analyst")

location = driver.find_element(By.XPATH, "/html/body/div/div[4]/div/div[2]/div[2]/div/form/div/div[1]/ul/li[2]/div/input")
location.send_keys('Bangalore')


searchjobs = driver.find_element(By.XPATH, "/html/body/div/div[4]/div/div[2]/div[2]/div/form/div/div[2]/div/button")
searchjobs.click()


job_title = []
location = []
company_name = []
experience_name = []

title = driver.find_elements(By.XPATH, '//div[@class="jobCard_jobCard__jjUmu  white-box-border jobCard"]/div/h2/a')
for i in title[0:10]:
    job_title.append(i.text)

loc = driver.find_elements(By.XPATH, '//div[@class="jobCard_jobCard__jjUmu  white-box-border jobCard"]/div/div[3]/div[2]')
for a in loc[0:10]:
    location.append(a.text)

company = driver.find_elements(By.XPATH, '//div[@class="jobCard_jobCard__jjUmu  white-box-border jobCard"]/div/div[2]/span')
for c in company[0:10]:
    company_name.append(c.text)
    

experience = driver.find_elements(By.XPATH, '//div[@class="jobCard_jobCard__jjUmu  white-box-border jobCard"]/div/div[3]/div[1]')
for d in experience[0:10]:
    experience_name.append(d.text)
    
df = pd.DataFrame({'title': job_title, 'location': location, 'company_name': company_name, 'experience_required': experience_name})
df   
```

```python
# 1.Fetch the webpage 

driver = webdriver.Chrome()
driver.get("https://www.azquotes.com/")

# 2.Enter the data analyst in the job title field, and Bangalore in the location field

search = driver.find_element(By.XPATH, "/html/body/div[1]/div[2]/div[1]/div/div[3]/ul/li[5]/a")
search.click()


quote = []
author = []
type_quote = []
    
for page in range(0,10):

    quote_list = driver.find_elements(By.XPATH, '//a[@class="title"]')
    for i in quote_list:
        quote.append(i.text)

    author_list = driver.find_elements(By.XPATH, '//div[@class="author"]/a')
    for a in author_list:
        author.append(a.text)

    type_list = driver.find_elements(By.XPATH, '//div[@class="tags"]')
    for c in type_list:
        type_quote.append(c.text)
        
    submit = driver.find_element(By.XPATH, '/html/body/div[1]/div[3]/div/div/div/div[1]/div/div[3]/li[12]/a')
    submit.click()
    
    time.sleep(4)    

df = pd.DataFrame({'Quote title': quote[0:1000], 'Author name': author[0:1000], 'Quote type': type_quote[0:1000]})
df   
```
